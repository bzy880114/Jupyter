```python
from random import choice

class RandWalk():
    
    def __init__(self, num_points=5000):
        self.num_points = num_points
        self.x_value = [0]
        self.y_value = [0]
    
    def get_step(self):
        direction = choice([1, -1])
        distance = choice([0, 1, 2, 3, 4, 5])
        return direction * distance
        
    def fill_walk(self):
        while len(self.x_value) < self.num_points:
            x_step = self.get_step()
            y_step = self.get_step()
            
            if (x_step == 0 and y_step == 0):
                continue
            
            next_x = self.x_value[-1] + x_step
            next_y = self.y_value[-1] + y_step
            
            self.x_value.append(next_x)
            self.y_value.append(next_y)
            
            
num_points = 2000
            
rw = RandWalk(num_points)
rw.fill_walk()

point_numbers = list(range(rw.num_points))
x_line = [0]
y_line = [0]

fig, axes = plt.subplots(1, 1, figsize=(20,20))
# axes.plot(rw.x_value, rw.y_value)
for i in range(num_points):
    if (i == 0):
        plt.scatter(rw.x_value[i], rw.y_value[i], s=200, c='purple', edgecolors='none')
    elif (i%50 == 0):
        plt.scatter(rw.x_value[i], rw.y_value[i], s=100, cmap=plt.cm.Greens, edgecolors='none')
        plt.annotate(int(i/50), xy = (rw.x_value[i], rw.y_value[i]), xytext = (rw.x_value[i]+1, rw.y_value[i]+1),  fontsize=12,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
        x_line.append(rw.x_value[i])
        y_line.append(rw.y_value[i])
    elif (i == num_points-1):
        plt.scatter(rw.x_value[i], rw.y_value[i], s=200, c='yellow', edgecolors='none')
        x_line.append(rw.x_value[i])
        y_line.append(rw.y_value[i])
plt.scatter(rw.x_value, rw.y_value, s=20, c=point_numbers, cmap=plt.cm.BrBG, edgecolors='none')
plt.plot(x_line, y_line, 'k-.', ls = ':')
plt.show()
```

```python
import requests as rs
```

```python
ulr = 'https://api.github.com/search/repositories?q=language:python&sort=stars'
r = rs.get(ulr)
print('Status code:', r.status_code)
```

```python
response_dict = r.json()
print(response_dict.keys())
```

```python
repo_dicts = response_dict['items']
print('Repositories returned:', len(repo_dicts))
```

```python
repo_dict = repo_dicts[0]
print('Keys:', len(repo_dict))
for key in sorted(repo_dict.keys()):
    print(key)
```

```python
repo_dict
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 20))
names, stars = [], []
for repo_dict in repo_dicts:
    names.append(repo_dict['name'][:10])
    stars.append(repo_dict['stargazers_count'])
    
ax.bar(names, stars)
ax.set_title('Programs') 
ax.set_xlabel('Program name')
ax.set_ylabel('Stars')
plt.xticks(names, rotation=315, )
plt.grid(axis='y')
plt.show()
```
